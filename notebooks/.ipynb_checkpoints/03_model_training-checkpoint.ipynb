```python
# === 1. Import Libraries ===
import pandas as pd
import numpy as np
import joblib # For saving our trained models

# --- Models ---
# We need two types: a Classifier (for win/loss) and a Regressor (for score difference).
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression # A good starting model for score prediction

# --- Evaluation Metrics ---
from sklearn.metrics import accuracy_score, classification_report # For the classification model
from sklearn.metrics import mean_absolute_error # For the regression model

# --- Plotting ---
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice plot style
sns.set_style('whitegrid')

print("Libraries imported successfully.")
```

```python
# === 2. Load Cleaned Dataset ===

# Define the path to your clean data
file_path = '../data/final/final_dataset.csv'

# Load the dataset using pandas
try:
    df = pd.read_csv(file_path)
    
    # IMPORTANT: Ensure the date column is a proper datetime object for sorting
    df['home_gameDate'] = pd.to_datetime(df['home_gameDate'])

    print(f"✅ Successfully loaded data from: {file_path}")
    print(f"Dataset shape: {df.shape}")

    # Display the first few rows to verify the data looks correct
    print("\nFirst 5 rows of the loaded dataset:")
    display(df.head())

except FileNotFoundError:
    print(f"❌ ERROR: File not found at '{file_path}'")
    print("Please make sure you have run the 02_feature_engineering notebook and saved the file.")
```

```python
# === 3. Split Data into Training and Testing Sets (Time-Based) ===

# We cannot shuffle time-series data. We must train on the past and test on the future.

# 1. Sort the entire dataset by date to ensure everything is in chronological order
# We will use the 'df' dataframe loaded in the previous step
df_sorted = df.sort_values(by='home_gameDate').reset_index(drop=True)
print("✅ Dataset sorted chronologically.")

# 2. Define a split point. We'll use the first 80% of games for training.
split_point = int(len(df_sorted) * 0.8)

# 3. Create the training and testing sets
train_data = df_sorted.iloc[:split_point]
test_data = df_sorted.iloc[split_point:]

print(f"\nTraining data (first 80% of games): {train_data.shape}")
print(f"  - Covers games from {train_data['home_gameDate'].min().date()} to {train_data['home_gameDate'].max().date()}")

print(f"\nTesting data (last 20% of games): {test_data.shape}")
print(f"  - Covers games from {test_data['home_gameDate'].min().date()} to {test_data['home_gameDate'].max().date()}")
```

```python
# === 4. Define Features and Targets (CORRECTED to prevent leakage) ===

# 1. Define the list of feature columns.
# We will ONLY use rolling averages calculated BEFORE the game.
# The 'diff_' columns contain post-game information and MUST be excluded.
feature_columns = [col for col in train_data.columns if '_rolling' in col]

# Fill any missing values with the median of the column.
X_train = train_data[feature_columns].fillna(train_data[feature_columns].median())
# Use the training data's median to fill the test set to prevent data leakage
X_test = test_data[feature_columns].fillna(train_data[feature_columns].median())

# 2. Define our TWO target variables (these are unchanged).
y_train_win = train_data['home_win']
y_test_win = test_data['home_win']
y_train_score = train_data['score_diff']
y_test_score = test_data['score_diff']


# --- Verification ---
print(f"✅ Found {len(feature_columns)} feature columns to use for training (Leakage Removed).")
print(f"\nTraining Features (X_train) Shape: {X_train.shape}")
print(f"Testing Features (X_test) Shape: {X_test.shape}")
print("\n✅ No missing values found in the training feature set.")

# Display the final list of columns being used as features
print("\nFinal feature columns selected:")
print(feature_columns)
```

```python
# === 5. Train and Evaluate Model 1: Win/Loss Prediction (Classification) ===

print("--- Training the Win/Loss Prediction Model ---")

# 1. Initialize the RandomForestClassifier model
win_loss_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# 2. Train the model using our training data
print("Training the model... (This may take a moment)")
win_loss_model.fit(X_train, y_train_win)
print("✅ Model training complete.")

# 3. Make predictions on the unseen test data
predictions_win = win_loss_model.predict(X_test)

# 4. Evaluate the model's performance
accuracy = accuracy_score(y_test_win, predictions_win)
print(f"\nModel Accuracy on Test Data: {accuracy:.4f}")
print("\nThis means the model correctly predicted the winner in approximately "
      f"{accuracy:.1%} of the games in the test set.")

# Display a more detailed report
print("\nClassification Report:")
print(classification_report(y_test_win, predictions_win))

# 5. Save the trained model to a file
# First, ensure the 'models' directory exists
import os
os.makedirs('../models', exist_ok=True)

model_filename = '../models/win_loss_classifier.pkl'
joblib.dump(win_loss_model, model_filename)

print(f"✅ Win/Loss prediction model saved to: {model_filename}")
```

```python
# === 6. Train and Evaluate Model 2: Score Difference Prediction (Regression) ===

print("--- Training the Score Difference Prediction Model ---")

# 1. Initialize a Linear Regression model
score_model = LinearRegression(n_jobs=-1)

# 2. Train the model using the same training data, but with the 'score_diff' target
print("Training the model...")
score_model.fit(X_train, y_train_score)
print("✅ Model training complete.")

# 3. Make predictions on the test data
predictions_score = score_model.predict(X_test)

# 4. Evaluate the model's performance
# For regression, we use metrics like Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test_score, predictions_score)

print(f"\nModel Mean Absolute Error (MAE) on Test Data: {mae:.4f}")
print("\nThis means that, on average, the model's score difference prediction "
      f"is off by about {mae:.2f} points.")

# 5. Save the trained regression model
model_filename = '../models/score_diff_regressor.pkl'
joblib.dump(score_model, model_filename)

print(f"\n✅ Score difference prediction model saved to: {model_filename}")
```

```python
# === 7. Analyze Feature Importance ===

# This shows us which features our Win/Loss model found most useful.

# 1. Get feature importances from the trained model
feature_importances = pd.Series(
    win_loss_model.feature_importances_, 
    index=X_train.columns
).sort_values(ascending=False)

# 2. Plot the results
plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances, y=feature_importances.index)
plt.title("Feature Importance for Win/Loss Prediction Model")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

# Display the top 5 most important features
print("Top 5 most important features:")
print(feature_importances.head(5))
```
